```python
import pandas as pd 
import folium
import numpy as np
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
calls = pd.read_csv('./data_files/911.csv')
calls
```

```python
call_unit = calls['title'].str.split(':', expand=True)
calls['call_unit'] = call_unit[0]
```

```python
calls = calls[:1500]
calls['call_unit'].unique()
```

```python
m = folium.Map(location=[np.mean(calls['lat']),np.mean(calls['lng'])], zoom_start=10)
car = folium.FeatureGroup(name='Traffic Incident', col='green')
ems = folium.FeatureGroup(name='EMS', col='blue')
fire = folium.FeatureGroup(name='Fire', col='red')

for index, info in calls.iterrows():
    if info['call_unit'] == 'Traffic':
        car_marker = folium.Marker([info['lat'], info['lng']], tooltip = f"{info['title']}, {info['timeStamp']}",icon=folium.Icon(prefix='fa',icon='car', color='green'))
        car.add_child(car_marker)
        m.add_child(car)



    if info['call_unit'] == 'EMS':
        ems_marker = folium.Marker([info['lat'], info['lng']], tooltip = f"{info['title']}, {info['timeStamp']}",icon=folium.Icon(prefix='fa',icon='ambulance'))
        ems.add_child(ems_marker)
        m.add_child(ems)



    if info['call_unit'] == 'Fire':
        fire_marker = folium.Marker([info['lat'], info['lng']], tooltip = f"{info['title']}, {info['timeStamp']}",icon=folium.Icon(prefix='fa',icon='fire-extinguisher', color='red'))
        fire.add_child(fire_marker)
        m.add_child(fire)
    
folium.map.LayerControl('topleft', collapsed= False).add_to(m) 


m
```
